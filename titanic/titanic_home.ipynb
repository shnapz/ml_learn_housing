```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer, StandardScaler, LabelBinarizer
from sklearn.base import BaseEstimator, TransformerMixin


class CabinExtractor(BaseEstimator, TransformerMixin):
    def __init__(self, cabin_col, enabled=True):
        self.enabled = enabled
        self.cabin_col = cabin_col
        self.blocks = {"A": 0,
                       "B": 1,
                       "C": 2,
                       "D": 3,
                       "E": 4,
                       "F": 5,
                       "G": 6,
                       "H": 7,
                       "K": 8,
                       "T": 9}
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        #target_except_given = np.delete(X, self.cabin_col, axis=1)
        #if not self.enabled:
        #    return target_except_given
        source = X
        block_col = np.empty((len(source), 11), dtype=int)
        #seat_col = np.empty((len(source), 11), dtype=int)
        block_col[:, :] = 0
        #seat_col[:, :] = np.nan
        for i in range(0, len(source)):
            x = source[i]
            if pd.isnull(x):
                block_col[i, 10] = 1
                continue
            splitted = x.split(" ")
            seats = np.empty((len(splitted), 2), dtype=object)
            for j in range(0, len(splitted)):
                cabin_seat = splitted[j]
                cabin_seat_number = cabin_seat[1:]
                if not cabin_seat_number:
                    cabin_seat_number = np.nan
                else:
                    cabin_seat_number = int(cabin_seat_number)
                seats[j] = [self.blocks[cabin_seat[0]], cabin_seat_number]
            for block in seats[:, 0]:
                block_col[i, block] = 1

        return block_col 
    

imputer = Imputer(strategy="median")
scaler = StandardScaler()
labelBin = LabelBinarizer()
cabin_extr = CabinExtractor(0, True)

# ["Name", "Sex", "Ticket", "Cabin", "Embarked"]

train_df = train_df.dropna(subset=["Embarked"])
#train_df["family"] = train_df["SibSp"] + train_df["Parch"] 

train_cat = train_df[["Sex"]].values

train_cabin = cabin_extr.fit_transform(train_df["Cabin"].values)

train_num = train_df[["Pclass", "Age", "SibSp", "Fare", "Parch"]].values
y_train = train_df["Survived"].values

train_num = imputer.fit_transform(train_num)
#train_num_scaled = scaler.fit_transform(train_num)


bined = labelBin.fit_transform(train_cat)

train_num_joined = np.concatenate((train_num, bined), axis=1)
train_num_joined2 = np.concatenate((scaler.fit_transform(train_num), bined, train_cabin), axis=1)
#train_num_joined2 = train_num_joined2)

train_num_joined2 = np.delete(train_num_joined2, 14, axis=1)
train_num_joined2 = np.delete(train_num_joined2, 13, axis=1)
train_num_joined2 = np.delete(train_num_joined2, 7, axis=1)


y = np.array(y_train.T)[np.newaxis]

pd.DataFrame(np.concatenate((y.T, train_num_joined2), axis=1)).corr()
```
